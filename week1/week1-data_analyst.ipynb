## What is Data? What is information?
### Data are measurable magnitudes collected or captured by some recording device. Information is data with context.

## What is Data Analytics?

### Data analytics is an interdisciplinary art of drawing meaningful insights i.e. information from stored datasets. Leadership, key stakeholders, and prospective clients can use these insights to make well-informed decisions. Data is present in all fields, domains, and practices.

## What is a Data Analyst and what are their typical responsibilities?

### A Data Analyst is someone who draws meaningful insights from data to describe trends or patterns to key stakeholders. The Data Analyst must have a robust and firm understanding of the domain space from which the data they analyze originates from.

### Here is a list of the more common business sectors that hire Data Analysts:
* Healthcare
* Finance/Budgeting
* Banking
* Retail

## What are the key characteristics of the ideal Data Analyst?
* Agile/Nimble/Quick to adjust mentally
* Inquisitive
* Scientific
* Methodical
* Creative

## What skills does the ideal Data Analyst possess?
* The ability to quickly learn new processes, techniques, and technologies
* The ability to quickly establish trust with key stakeholders or any audience of findings
* The ability to listen intently to draw business requirements, logic, and prior work performed on project
* The ability to eloquently explain their findings to any audience
* The ability to visualize data using the optimal design choices
* Great interpersonal skills with clients and team members
* The understanding of how to use the resources available to them in an optimal way
* The understanding of where he/she/they exist in the tech stack - know who to go for help and how the project as a whole works

## What is Data Science and what is the difference between a Data Scientist and a Data Analyst?
### Data Science is a broader discipline than Data Analytics that is usually focused on developing or researching novel methods via statistics, complex algorithms, and new data to provide insights helpful towards creating new capabilities using data. Data Analytics is more focused on using past data to generate strategies and planning for further improvement or development of a business, firm, or university.
