## What does it mean to program at the high level?
### Programmers write instructions to a mechanical computer to execute using either an application or software, compiled language, or interpreted language. Programming at the high level refers to programmers who write instructions to interpreters or compilers that abstract the instructions for the hardware to operate on (i.e. the motherboard)

### Application and software examples:
* Host OS (Windows, MacOS, Linux)
* MSFT Suite (Word, PPTX, Excel)

### Compiled language uses a compiler that assembles your human language instructions and converts them into machine language for the mechanical computer to understand (i.e. binary). 

### Interpreted languages use an interpreter to load the human language instructions into a format which then are ran against the a virtual machine. In Python's case, the code is formatted into bytecode which is then ran against the Python Virtual Machine.

## What tools and conditions are required to use Python?
### Python is virtually accessible to anyone with a mechanical computer and wi-fi connection. However, for seasoned professionals, they use their own hardware or account's cloud-provisioned hardware, Python installation, and often but not required Integrated Development Environment(s) or IDE

## What is an IDE?
### An IDE is an application with added features to assist the programmer to write code in their programming language of choice.

## Why is Python such a popular programming language?
### Python has an extremely active community of users who contribute support to the programming language by allowing other programmers to use their own code enabling them to perform tasks easier and more efficiently. Python also lacks many programming conventions such as declarations and linkages to operate making it easier to use and thus more accessible than other languages.

### Other reasons:
* Flexible
* Easy-to-understand syntax
* Intuitive and easy to use
* Reliable
* Well-supported and well-documented
* Open-source
